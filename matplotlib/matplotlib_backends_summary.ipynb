# Summary - Matplotlib Backends

**Purpose**: Determine how well different matplotlib backends work for the following 3 cases when used from an IPython notebook:
1. Simple plotting (either inline or in a new window)
2. Plot animation with FuncAnimation
3. Plot manipulation with the interact IPython widget

The two tables below summarize the results of my testing on a laptop running Mac OS X 10.9.4 and Windows 7 (the latter inside a VMWare virtual machine). In both cases, I am using an Anaconda Python 2.7.x installation.

**Mac OS X**  

backend            | fig location for simple plt.plot() | plot animation | interact 
:-----:            |:------------:|:-------:|:--------:
*MacOSX* (default) | Yes: new window  | No: animation & fails with error      | No: sliders inline but no plot anywhere
*Qt4Agg*           | Yes: new window | No: in new window right panel animates, but left doesn't | No: sliders inline but no plot anywhere
*tkagg*            | No: no window/plot | **Yes: works great in new window**    | No: sliders inline but no plot anywhere
*nbAgg*            | **Yes: inline**      | No: plot shows up inline but doesn't animate | No: sliders inline but no plot anywhere
*inline*           | **Yes: inline**      | No: one blank inline plot shows up      | **Yes: works fine**

**Windows 7**  

backend           | fig location for simple plt.plot() | plot animation | interact 
:-----:           |:------------:|:-------:|:--------:
*Qt4Agg* (default)| Yes: new window | **Yes-works great in new window** | No: sliders inline but no plot anywhere
*tkagg*           | Yes: new window | **Yes-works great in new window** | No: sliders inline but no plot anywhere
*nbAgg*           | **Yes: inline**   | No: plot shows up inline but doesn't animate | No: sliders inline but no plot anywhere
*inline*          | **Yes: inline** | No: one blank inline plot shows up | **Yes-works fine**

Note: plot animation uses FuncAnimation with corrected code from http://stackoverflow.com/questions/20016463/matplotlib-funcanimation-slow?rq=1 and a reduced number of time steps to limit how long the animation runs.

# How to change backends

There are two methods, *matplotlib.use('desired backend')* or *plt.switch_backend('desired backend')*. However, the former can only be used before other modules are imported and is therefore less flexible. Consequently, we will use only on the 2nd method.

Note that capitalization of the text string specifying each backend doesn't matter.

```python
# Example

plt.switch_backend('Qt4Agg')
plt.get_backend()
```

**Note**: In all of the following examples, restart the kernel before execution. This is not strictly necessary in most cases, but it is in a few so you will save yourself some trouble by following this advice.

# Try simple plot with plt.plot(x,y) for different backends

**Restart kernel now**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,4)
y = np.sin(2.0*np.pi*x)

plt.get_backend()
```

### MacOSX Backend

```python
# This creates a plot in a new window. The cell label will remain "In [*]" until the new window is closed,
# after which it becomes "In [NUM]"
plt.plot(x,y)
plt.show()
```

### Qt4Agg Backend

```python
plt.switch_backend('qt4agg')
plt.get_backend()
```

```python
# This creates a plot in a new window. The cell label will remain "In [*]" until the new window is closed,
# after which it becomes "In [NUM]". Note that the new window on Mac OS X is behind the notebook window
# in the center of the screen.
plt.plot(x,y)
plt.show()
```

### nbAgg Backend

```python
plt.switch_backend('nbagg')
plt.get_backend()
```

```python
# With the nbAgg backend, the plot is shown inline
plt.plot(x,y)
plt.show()
```

### inline Backend

```python
%matplotlib inline
plt.get_backend()
```

```python
plt.plot(x,y)
```

```python
# Define a simple pulse function. Let's use 1/2 cycle of a cosine centered at 0, and zero everywhere else
def pulse(a):
    if a < np.pi/2.0 and a > -np.pi/2.0:
        return np.cos(a)
    else:
        return 0.0

# Create a function to plot a forward propagating pulse. Use interact() to make it interactive.
zmin = -30
zmax = 30
numpnts = 500
def plotpulse(u,t):
    x = np.linspace(zmin,zmax,numpnts)
    y = np.zeros(numpnts)
    for i in range(0,numpnts): 
        y[i] = pulse(x[i] - u*t)
    plt.plot(x,y)
    plt.ylim(0,1.2)
    plt.xlabel('z')
    plt.figtext(0.15,0.82,'t = ' + str(t))
interact(plotpulse, u=(0,2,0.25), t=(-30,30,0.25));
```

# Try FuncAnimation with different backends

Read the comments carefully for instructions.

**Resart kernel now**

```python
# Corrected version from From http://stackoverflow.com/questions/20016463/matplotlib-funcanimation-slow?rq=1
# Use smaller number of time steps & play with things

import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation
from matplotlib.collections import PatchCollection
from random import *

# UNCOMMENT THE LINE BELOW FOR THE BACKEND YOU WANT TO TRY, THEN RESTART THE KERNEL AND EXECUTE THIS INPUT CELL
# If you leave all of the lines commented, the default backend will be used depending on your operating system

#plt.switch_backend('qt4agg')
plt.switch_backend('tkagg')
#plt.switch_backend('nbagg')
#%matplotlib inline


nx = 20     # x size of lattice
ny = 20     # y size of lattice

pAds = 0.01     # adsorption probability per time step
pDes = 0.0075   # desorption probability per time step

tMax = 50     # number of time steps

surface = np.zeros((nx,ny))              # create surface
xc = [0]
yc = [0]


# initialization and time step of simulation

patches = []

def init():
    global patches
    if patches:
        # prevent the second call of the init()
        return patches
    # initialize an empty list of circles
    for x in range(nx):
        for y in range(ny):
            if(surface[x][y] == 0):
                patches.append(ax_surf.add_patch(plt.Circle((x+0.5,y+0.5),0.45,color='w')))
    lines, = ax_covr.plot([],[])
    patches.append(lines)
    return patches

def animate(i): 
    global patches
    idx = 0
    for x in range(nx):
        for y in range(ny):
            if surface[x][y] == 0:
                if random() < pAds:
                    surface[x][y] = 1
                    patches[idx] = ax_surf.add_patch(plt.Circle((x+0.5,y+0.5),0.45,color='b'))
            else:
                if(random()<pDes):
                    surface[x][y] = 0
                    patches[idx] = ax_surf.add_patch(plt.Circle((x+0.5,y+0.5),0.45,color='w'))
            idx += 1
    coverage = np.sum(surface)/(nx*ny)
    xc.append(i)
    yc.append(coverage)
    lines, = ax_covr.plot(xc,yc,'ro',ms=2,lw=0)
    patches[idx] = lines
    return patches

# set up figure and animate

fig = plt.figure()
ax_surf = plt.subplot2grid((1, 2), (0, 0))
ax_covr = plt.subplot2grid((1, 2), (0, 1))
ax_surf.set_xlim(0,nx)
ax_surf.set_ylim(0,ny)
ax_covr.set_xlim(0,tMax)
ax_covr.set_ylim(0,1)

ax_surf.set_aspect(1)
ax_surf.axis('off')

ax_covr.set_aspect(tMax)

ax_surf.hold(False)

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=tMax, interval=0, blit=True,repeat=False)
plt.show()
```
